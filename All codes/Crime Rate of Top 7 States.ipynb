```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

file_path = "D:\\Anindya\\ISI Project - Data Cleaning + Data Annotation\\Crime Rate.csv"  # Update this


df = pd.read_csv(file_path)


df.head()
```

```python

states_of_interest = ['Kerala', 'Maharashtra', 'Madhya Pradesh', 'Tamil Nadu', 'Gujarat', 'Uttar Pradesh', 'Rajasthan']  # Replace with your actual state names


df_filtered = df[df['State'].isin(states_of_interest)]


df_filtered.head()
```

```python

sns.set(style="whitegrid")
```

```python

for state in states_of_interest:
    plt.figure(figsize=(8, 5))  
    
    
    state_data = df_filtered[df_filtered['State'] == state]
    
    
    state_data_grouped = state_data.groupby('Year')['Crime Rate'].mean().reset_index()
    
    
    sns.lineplot(x='Year', y='Crime Rate', data=state_data_grouped, marker='o')
    
    
    plt.title(f'Crime Rate in {state} (2018-2022)', fontsize=16)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Crime Rate (number of crimes per 1,000 people)', fontsize=12)
    plt.xticks(ticks=[2018, 2019, 2020, 2021, 2022])  # X-axis values
    plt.grid(True)
    
    
    plt.tight_layout()
    plt.savefig(f'crime_rate_{state}.png', bbox_inches='tight') 
    plt.show()
```
